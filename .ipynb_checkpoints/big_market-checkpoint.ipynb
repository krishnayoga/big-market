```python
##### Cek NaN data test #####

test_data[test_data.isnull().any(axis=1)]

g = test_data.count()
print(g)
print(test_data.dtypes)

h = test_data['Item_Weight'].value_counts()
print(h)
test_data = test_data.fillna({'Item_Weight':10.50})

i = test_data['Outlet_Size'].value_counts()
print(i)
test_data = test_data.fillna({'Outlet_Size':'Medium'})
```

```python
##### encode the train data ######
print(test_data.dtypes)

gb = pd.get_dummies(test_data, columns=['Item_Fat_Content','Item_Type','Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type'])
print(gb)
print(gb.dtypes)

x_val_all = gb.copy()
x_val = x_val_all.drop(columns=['Item_Identifier'])
print(x_val.shape)

x_val_np = x_val.to_numpy().astype('float64')
# x_val_norm = normalize(x_val_np)
# print(x_val_norm.shape)

x_val_scaled = scaler.fit_transform(x_val_np)
print(x_val_scaled.shape)
```

```python
##### separate x and y, train and test

#### data fb

x_all = fb.copy()
# print(x_all)

x = x_all.drop(columns=['Item_Identifier','Item_Outlet_Sales'])
y = x_all['Item_Outlet_Sales']
# print(x)
# print(y)

# ##### convert to numpy
x_np = x.to_numpy().astype('float64')
y_np = y.to_numpy().astype('float64')
print(x_np)
print(y_np)

y_np.resize((len(y_np),1))
print(y_np)

# x_np_norm = normalize(x_np)
# y_np_norm = normalize(y_np)
# print(x_np_norm)
# print(y_np_norm)

# scaler1 = MinMaxScaler()


x_np_scaled = scaler.fit_transform(x_np)
y_np_scaled = scaler.fit_transform(y_np)

print(x_np_scaled)
print(y_np_scaled)

##### split into x_tr, x_te, y_tr, y_te
# x_train, x_test, y_train, y_test = train_test_split(x_np_norm,y_np_norm,train_size=0.90)
x_train, x_test, y_train, y_test = train_test_split(x_np_scaled,y_np_scaled,train_size=0.90)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
print(y_test)
```

```python
model = keras.Sequential(
    [
        layers.Dense(128, activation="relu", input_shape=(x_train.shape[-1],)),
        layers.Dense(256, activation="relu"),
        layers.Dense(128, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dense(1, activation="relu"),
    ]
)
model.summary()

model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.05, momentum=0.04, nesterov=False, name="SGD"), 
              loss=keras.losses.MeanSquaredError(),
             metrics=[keras.metrics.SparseCategoricalAccuracy(name="acc")])

model.fit(x_train,y_train,batch_size=767,epochs=100)

print('val')
model.fit(x_test,y_test,batch_size=853,epochs=1)

model.layers[2].output.get_shape()
```

```python
get_3rd_layer_output = K.function([model.layers[0].input],
                                  [model.layers[4].output])
layer_output = get_3rd_layer_output([x_val_scaled])[0]

layer_output = scaler.inverse_transform(layer_output)
print(layer_output.shape)
print(layer_output)
# scaler.inverse_transform(layer_output)

with open("result.csv", 'w', newline='') as f:
    csv.writer(f).writerows(layer_output)
```
